```python
%%time
import pandas as pd
import numpy as np
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from itertools import combinations 
import seaborn as sns; sns.set(style="ticks", color_codes=True)
```

```python
%%time
train = pd.read_csv("1_min_train.csv")
test = pd.read_csv("1_min_test.csv")
```

```python
print(train.shape)
print(test.shape)
```

```python
features=train.columns.tolist()
features = features[3:]
features
```

# Combination general funtion

```python
def get_combination(arr,n): 
    return list(combinations(arr, n))
```

# Combination of 2

```python
comb = get_combination(features,2)
i = 1
for c in comb:
    print (c)
    print (str(i) +" out of "+str(len(comb)))
    
    et = ExtraTreesClassifier(n_estimators=50, n_jobs=10, verbose=1,random_state=0)
    
    et.fit(train[list(c)],train['target'])
    
    y_pred=et.predict(test[list(c)])
    
    print(classification_report(test['target'], y_pred))
    i = i+1
```

# Feature Importance

```python
et = ExtraTreesClassifier(n_estimators=20, n_jobs=10, verbose=2,random_state=0)
et.fit(train[features],train['target'])
y_pred=et.predict(test[features])
print(classification_report(test['target'], y_pred))
```

```python
feat_importances = pd.Series(et.feature_importances_, index=train[features].columns)
feat_importances.plot(kind='barh')
feat_importances
```
