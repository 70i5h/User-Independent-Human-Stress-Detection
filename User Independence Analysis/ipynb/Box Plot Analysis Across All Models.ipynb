```python
def calc_z(percentage):
    if (percentage==0.05):
        z=(((.1/398)*148))
    if (percentage==.1):
        z=(((.1/395)*102)+.1)
    if (percentage==.2):
        z=(((.1/386)*207)+.2)
    if (percentage==.3):
        z=(((.1/375)*321)+.3)
    if (percentage==.4):
        z=(((.1/342)*85)+.5)
    if (percentage==.5):
        z=(((.1/323)*243)+.6)
    if (percentage==.6):
        z=(((.1/278)*119)+.8)
    if (percentage==.7):
        z=(((.1/230)*87)+1.0)
    if (percentage==.8):
        z=(((.1/183)*151)+1.2)
    if (percentage==.9):
        z=(((.1/102)*48)+1.6)
    if (percentage==.95):
        z=(((.1/59)*22)+1.9)
    if (percentage==1):
        z=3
    #print(z)
    return(z)
    
def excluding_outlayers(mean,std):

    for i in [0.05,.1,.2,.3,.4,.5,.6,.7,.8,.9,.95,1]:
        #print('for',int(i*100),'%',)
        z=calc_z(i)
        
        pt1 = mean + z *std

        pt2 = mean - z *std
        print(str(round(((pt1+pt2)/2),3))+'±'+str(round(((pt1-pt2)/2),3)))
cls=[2,3,4]
cls_string=['Bi-affective state','Tri-affective state','Multi-affective state']
spt=globals()['%s_label_2_class' % name[0]]
for i in range(3):
    for j in range(3,7):


        mean = globals()['Random_Forest_mean_%s_class' % cls[i]][j]
        std = globals()['Random_Forest_std_%s_class' % cls[i]][j]
        
        #print(i,j)
        print('Random_Forest',cls_string[i],'split',spt[j],'mean',mean,'std',std)
        excluding_outlayers(mean,std)
```
