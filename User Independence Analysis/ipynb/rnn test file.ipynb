```python
def one_hot(y, numOutput):
    print(len(y))
    y = np.reshape(y, [-1])
    print(len(y))
    ret = np.zeros([y.shape[0], numOutput])
    print(len(ret))
    for i, label in enumerate(y):
        ret[i, label] = 1
    return ret
```

```python
from sklearn.model_selection import train_test_split
import pathlib

xxxx = [some[0] for some in y_train]
y_train=np.array(xxxx)

xxxx = [some[0] for some in y_test]
y_test=np.array(xxxx)

timesteps = x_train.shape[-2] #128/256
feats = x_train.shape[-1]  #16

trainSize = int(x_train.shape[0]*0.9) #6566
x_train, x_val = x_train[:trainSize], x_train[trainSize:] 
y_train, y_val = y_train[:trainSize], y_train[trainSize:]
print(x_val.shape)
print(y_val.shape)
print(len(x_val))
print(len(y_val))
```

```python
# normalization
x_train = np.reshape(x_train, [-1, feats])
mean = np.mean(x_train, axis=0)
std = np.std(x_train, axis=0)

# normalize train
x_train = x_train - mean
x_train = x_train / std
x_train = np.reshape(x_train, [-1, timesteps, feats])

# normalize val
x_val = np.reshape(x_val, [-1, feats])
x_val = x_val - mean
x_val = x_val / std
x_val = np.reshape(x_val, [-1, timesteps, feats])

# normalize test
x_test = np.reshape(x_test, [-1, feats])
x_test = x_test - mean
x_test = x_test / std
x_test = np.reshape(x_test, [-1, timesteps, feats])

# shuffle test, as this was remaining
idx = np.arange(len(x_test))
np.random.shuffle(idx)
x_test = x_test[idx]
y_test = y_test[idx]


print(len(x_val))
print(len(y_val))
```

```python
def create_model(input_length):
    print ('Creating model...')
    model = Sequential()
#     model.add(Embedding(input_dim = 200, output_dim = 50, input_length = input_length))
    model.add(LSTM(output_dim=256,input_shape=(628,17), activation='sigmoid', inner_activation='hard_sigmoid', return_sequences=True))
    model.add(Dropout(0.5))
    model.add(LSTM(output_dim=256, activation='sigmoid', inner_activation='hard_sigmoid'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))

    print ('Compiling...')
    model.compile(loss='binary_crossentropy',
                  optimizer='rmsprop',
                  metrics=['accuracy'])
    return model
```
