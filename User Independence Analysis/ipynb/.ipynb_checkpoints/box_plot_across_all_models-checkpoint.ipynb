```python
def diff(list_1,list_2):
    list_3=[]
    for i in range(len(list_1)):
        list_3.append(list_1[i]-list_2[i])
    return list_3
def delta(t):
    ret = [j-i for i, j in zip(t[:-1], t[1:])]
    return ret
```
