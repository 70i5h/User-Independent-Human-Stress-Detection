```python
features=df_new_1.columns.tolist()
features

removed = ['label']
for rem in removed:
    features.remove(rem)
features_with_sub=[]
features_with_sub[:]=features
removed = ['subject']
for rem in removed:
    features.remove(rem)

feature=features
print(len(feature))
len(features_with_sub)
```

```python
train=df_new_1[df_new_1['subject']<=9]
test=df_new_1[df_new_1['subject']>9]
```

```python
scaler = Normalizer()
scaled_data_train = scaler.fit_transform(train[feature])
scaled_data_test = scaler.transform(test[feature])
```

```python
et = ExtraTreesClassifier(n_estimators=50,n_jobs=10,random_state=237)
et.fit(scaled_data_train,train['label'])
y_pred=et.predict(scaled_data_test)
```

```python
print(classification_report(test['label'],y_pred))
```

```python
feature_importances = pd.DataFrame(et.feature_importances_,index = feature,columns=['importance']).sort_values('importance', ascending=False)
```

```python
feature_importances
```

# Feature Selection with feature_importances


```python
all_features = feature_importances.index
```

```python
temp = []
for i in all_features:
    temp.append(i)
#     print (temp)
    scalerr = Normalizer()
    scaled_data_trainn = scalerr.fit_transform(train[temp])
    scaled_data_testt = scalerr.transform(test[temp])
    
    ett = ExtraTreesClassifier(n_estimators=50,n_jobs=10,random_state=237)
    ett.fit(scaled_data_trainn,train['label'])
    y_pred=ett.predict(scaled_data_testt)
    
    rpt = classification_report(test['label'],y_pred,output_dict=True)['accuracy']
    
    if rpt >= 0.75:
        print (len(temp))
        print (classification_report(test['label'],y_pred))
        
    
```

# Feature Selection with feature_importances using forward Selection


```python
all_features_c = copy.deepcopy(all_features)
mas = []

while len(all_features_c)!=0:
    temp = []
    acc_list = []
    all_features_c = [x for x in all_features_c if x not in mas]
    print ("**************************",len(all_features_c))

    for i in all_features_c:
        temp.append(i)
        temp = temp + mas
#         print (temp)
        scalerr = Normalizer()
        scaled_data_trainn = scalerr.fit_transform(train[temp])
        scaled_data_testt = scalerr.transform(test[temp])

        ett = ExtraTreesClassifier(n_estimators=100,n_jobs=10,random_state=56)
        ett.fit(scaled_data_trainn,train['label'])
        y_pred=ett.predict(scaled_data_testt)

        acc_list.append(classification_report(test['label'],y_pred,output_dict=True)['accuracy'])

    ind = acc_list.index(min(acc_list))
    val = all_features_c[ind]
    mas.append(val)
    all_features_c.remove(val)
    print ("**************************",mas,min(acc_list))
```
