```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('personal_detail.csv',index_col=0)
df
```

```python
for i in df:
    print(df[i].unique())
    print(len(df[i].unique()))
```

```python
grouped=df.groupby(['gender_ female','coffee_today_YES','sport_today_YES','smoker_YES','feel_ill_today_YES'], as_index=False).agg({'subject':'count','height': ['mean','min', 'max',],'age': ['mean','min', 'max',],'weight': ['mean','min', 'max',]})
grouped
```

```python
for i in grouped:
    print(i[0])
```

```python
grouped['subject']['count']
```

```python
gender=list(grouped['gender_ female'].replace({0: 'Male', 1: 'Female'}))
gender
```

```python
coffee=list(grouped['coffee_today_YES'].replace({0: 'No', 1: 'Yes'}))
coffee
```

```python
sport=list(grouped['sport_today_YES'].replace({0: 'No', 1: 'Yes'}))
sport
```

```python
smoker=list(grouped['smoker_YES'].replace({0: 'No', 1: 'Yes'}))
smoker
```

```python
feel_ill=list(grouped['feel_ill_today_YES'].replace({0: 'No', 1: 'Yes'}))
feel_ill
```

```python
total_subject=list(grouped['subject']['count'])
total_subject
```

```python
height_range=[]
for i in range(len(list(grouped['height']['mean']))):
    val=list(grouped['height']['max'])[i]-list(grouped['height']['min'])[i]
    #print(val)
    if (val>0):
        height_range.append(str(list(grouped['height']['min'])[i])+"-"+str(list(grouped['height']['max'])[i]))
    else:
        height_range.append(str(list(grouped['height']['max'])[i]))
height_range
```

```python
age_range=[]
for i in range(len(list(grouped['age']['mean']))):
    val=list(grouped['age']['max'])[i]-list(grouped['age']['min'])[i]
    #print(val)
    if (val>0):
        age_range.append(str(list(grouped['age']['min'])[i])+"-"+str(list(grouped['age']['max'])[i]))
    else:
        age_range.append(str(list(grouped['age']['max'])[i]))
age_range
```

```python
weight_range=[]
for i in range(len(list(grouped['weight']['mean']))):
    val=list(grouped['weight']['max'])[i]-list(grouped['weight']['min'])[i]
    #print(val)
    if (val>0):
        weight_range.append(str(list(grouped['weight']['min'])[i])+"-"+str(list(grouped['weight']['max'])[i]))
    else:
        weight_range.append(str(list(grouped['weight']['max'])[i]))
weight_range
```

```python
df_new=pd.DataFrame([gender,coffee,sport,smoker,feel_ill,total_subject,height_range,age_range,weight_range]).T
df_new.columns=['Gender','Coffee','Sports','Smoked','Feel_ill','No_of_subject','Height','Age','Weight']
df_new
```

```python
df_new.to_csv('user_detail.csv')
```
